## 1.1 Function, limits, continuity

#### Function: 

- A rule that associates pairs of numbers. 

- A function is a relationship between two sets of data, one input and the other output. A relation is said to be a function iff each of the input data is related to exactly one output data.


##### Properties of a function:
- The domain of a function f corresponds to the set of values that its independent variable, generally x, can take.
- The range of a function f corresponds to the set of values the dependent variable, generally y, can take.
- A function is strictly increasing over an interval of x if and only if the values of y are always increasing.
- A function is strictly decreasing over an interval of x if and only if the values of y are always decreasing.

##### Limits:
- evaluate the function at values arbitrarily close to a chosen input value, without actually reaching that value.

##### Continuity:
- a function is continuous and smooth iff the following rules are met:
    - $f(a)$ is defined, tha is value of function at a,
    - limit of $f(x)$ as x approaches a exists,
    - the value of the function at a is the same as the limit.

```python
# define a function, say f(x) = (x**3-1)/(x-1)
def f(x):
    return (x**3-1)/(x-1)
```

```python
# calculate value of f(x) for some values and plot

import matplotlib.pyplot as plt

for i in range(-10, 10):
    if i != 1:
        plt.plot(i, f(i), '.')
plt.show()
```

## 1.2 Calculus - Derivative

##### Calculating rate of change

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline
```

```python
# define a function such as f(x) = x**3
def f(x):
    return x**3
```

```python
# define derivative function
def derivative(x):
    # set delta (change in x) to a small value, say 0.001
    delta = 1.0/1000.0
    # define rise (the change in f(x))
    rise = (f(x + delta)) - f(x)
    slope = rise / delta
    return slope
```

```python
# test the derivative function for some values of x
print('derivative of f(2) = ', derivative(2))
print('derivative of f(-7) = ', derivative(-7))
```

```python
# plot derivative of f(x)
x = np.arange(-10,11,1)
y = derivative(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('derivative(x)')
plt.grid()
plt.show()
```

## 1.3 Calculus - Integral

##### To estimate the area under the curve. If function is a graph, integration helps you find the total area under the curve between two points.

$\int_{a}^{b} f(x)dx$

Applications - probability distributions and area under the curve estimation

```python
import sympy as sp

def calculate_integral(start, end):
    # Define the symbol x
    x = sp.symbols('x')
    
    # Define the function 2x
    function = 2 * x
    
    # Compute the definite integral of the function
    integral_value = sp.integrate(function, (x, start, end))
    
    return integral_value

# Call the function and print the result
print(calculate_integral(0, 1))
```
